# DB SCAN

```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
X, y = make_moons(n_samples=500, noise=0.1)
```

```python
plt.scatter(X[:,0], X[:,1], c=y, label=y)
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
```

```python
#menggunakan parameter default

dbscan_cluster1 = DBSCAN()
dbscan_cluster1.fit(X)
plt.scatter(X[:,0],X[:,1],c=dbscan_cluster1.labels_,label=y)
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
```

```python
#mengubah parameter (nilai epsilon 0.2 dan minimum Points=5)
dbscan_cluster2 = DBSCAN (eps=0.2, min_samples=5)
dbscan_cluster2.fit(X)
plt.scatter(X[:, 0], X[:, 1],
c=dbscan_cluster2.labels_, label=y)
plt.xlabel("$x_15")
plt.ylabel("$x 25")
```
